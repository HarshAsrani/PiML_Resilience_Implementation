```python
import pandas as pd
from resilience_testing import Experiment
from xgboost import XGBRegressor
```

```python
data = pd.read_csv('BikeSharing.csv')
experiment = Experiment(data, 'regression', 'cnt', exclude=['mnth', 'yr', 'temp'])
model = XGBRegressor(n_estimators=100, max_depth=2)
experiment.model_train(model, "XGBoost")
```

```python
experiment.model_diagnose(show='resilience_perf', resilience_method="worst-sample")
```

```python
experiment.model_diagnose(show="resilience_perf", resilience_method="outer-sample")
```

```python
experiment.model_diagnose(show="resilience_perf", resilience_method="worst-cluster")
```

```python
experiment.model_diagnose(show="resilience_perf", resilience_method="hard-sample")
```

```python
experiment.model_diagnose(show='resilience_distance', resilience_method='worst-sample', distance_metric='psi')
```

```python
experiment.model_diagnose(show="resilience_distance", resilience_method="worst-sample",
                         distance_metric="psi", immu_feature="hr", alpha=0.3)
```

```python
experiment.model_diagnose(show="resilience_distance", resilience_method="worst-cluster",
                   distance_metric="wd1", n_clusters=10)
```
